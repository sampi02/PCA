# PCA Wine

## Perform Principal component analysis and perform clustering using first 3 principal component scores (both heirarchial and k mean clustering(scree plot or elbow curve) and obtain optimum number of clusters and check whether we have obtained same number of clusters with the original data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine = pd.read_csv("C:\\Users\\User\\Documents\\Datascience_Assignments\\Assignment 8 (PCA)\\wine.csv")
wine
```

```python
wine['Type'].value_counts()
```

```python
wine2 = wine.iloc[:,1:]
wine2
```

```python
wine2.isnull().sum()
```

```python
wine_ary = wine2.values
wine_ary
```

```python
wine_norm = scale(wine_ary)
wine_norm
```

## PCA Implementation

```python
pca = PCA(n_components=13)
wine_pca = pca.fit_transform(wine_norm)
wine_pca
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1,color='purple')
```

```python
x = wine_pca[:,0:1]
y = wine_pca[:,1:2]
z = wine_pca[:,2:3]
plt.scatter(x,y)
plt.scatter(y,z)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,'g*')
plt.title('PC1&2')
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(y,z,'r^')
plt.grid('True')
plt.title('PC2&3')
plt.show()
```

```python
final_df = pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

### Visualization of PCAs

```python
fig = plt.figure(figsize = (16,12))
sns.scatterplot(data = final_df)
```

# Checking with other Clustering Algorithms

### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hclusters = AgglomerativeClustering(n_clusters = 3,affinity = 'euclidean',linkage = 'ward')
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(wine_norm),columns = ['clustersid'])
y['clustersid'].value_counts()
```

```python
wine3 = wine.copy()
wine3['clustersid'] = hclusters.labels_
wine3
```

### K-Means Clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Build Cluster algorithm using K=3

```python
clusters3 = KMeans(3, random_state = 30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
wine4 = wine.copy()
wine4['clusters3id'] = clusters3.labels_
wine4
```
